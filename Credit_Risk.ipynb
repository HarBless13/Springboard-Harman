## Data Cleaning

```python
# to check the datatyes in the dataset
df.dtypes
```

```python
# tells about information of each column of data
df.info()
```

```python
# Convert column names to keep first letter captal while other letters lowercase and keep the underscores
df.columns = df.columns.str.replace("_", " ").str.title().str.replace(" ", "_")
```

```python
df.head()
```

```python
# To see how many missing values in each column
df.isna().sum()
```

```python
## to check duplicate data
df.duplicated().sum()
```

```python
## To display duplicate values and calculate its percentage
duplicate_rows = df[df.duplicated(keep='first')].copy()
duplicate_rows.fillna(0, inplace=True)

# number of duplicates
count = len(duplicate_rows)
# total number of rows
rows = len(df)
# percentage
per = (count / rows) * 100
# Print the percentage
print(f"Percentage of Datasets that are duplicates: {per:.2f}%")

# Display duplicate rows
duplicate_rows
```

##### 0.51% duplicate rows in the dataset are identified. Duplicates were defined as rows having identical values across all features. Since duplicates do not provide additional information and may lead to biased model evaluation if present in both training and test sets, they were removed. This step resulted in a cleaner dataset without significant loss of data.

```python
# to remove duplicates
df = df.drop_duplicates(keep='first')
print(f"Shape after removing duplicates: {df.shape}")
```

```python
df['Loan_Status']
```

```python
# Convert Cb_Person_Default_On_File's value into binary
df['Cb_Person_Default_On_File'] = df['Cb_Person_Default_On_File'].replace({'Y': 1, 'N': 0})

# Fill missing values with most common(mode)
df['Cb_Person_Default_On_File'].fillna(df['Cb_Person_Default_On_File'].mode()[0], inplace=True)
df['Cb_Person_Default_On_File'] = df['Cb_Person_Default_On_File'].astype(int)


print(df.dtypes)
```

```python
df.isna().sum()
```

```python
# to check the unique values in the columns
print(df['Person_Emp_Length'].unique())
print(df['Loan_Int_Rate'].unique())
```

```python
# Handle missing values in Person_Emp_Length
df['Person_Emp_Length'].fillna(df['Person_Emp_Length'].median(), inplace=True)
```

```python
# Handle missing values in Loan_Int_Rate
df['Loan_Int_Rate'] = df.groupby('Loan_Grade')['Loan_Int_Rate'].transform(
    lambda x: x.fillna(x.median())
)
# print the number of missing values after filling
print("Missing Loan_Int_Rate AFTER filling:", df['Loan_Int_Rate'].isna().sum())
```

```python
df.isna().sum()
```

```python
# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

# Save cleaned dataset
df.to_csv("credit_risk_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```

## Exploratory Data Analysis (EDA)

```python
# Load cleaned dataset
df = pd.read_csv("credit_risk_cleaned.csv")
df.head()
```

```python
# Data types and non-null counts
df.info()

# Summary statistics (numeric columns)
df.describe()

# Summary statistics (categorical columns)
df.describe(include=['object'])
```

```python
# Distribution of Loan_Status (target variable)
sns.countplot(x="Loan_Status", data=df)
plt.title("Loan Approval Status Distribution")
plt.show()

print(df['Loan_Status'].value_counts(normalize=True) * 100)
```

```python
# Univariate Analysis
# Numeric features
numeric_cols = ['Person_Age', 'Person_Income', 'Person_Emp_Length',
                'Loan_Amnt', 'Loan_Int_Rate', 'Loan_Percent_Income',
                'Cb_Person_Cred_Hist_Length']

df[numeric_cols].hist(bins=30, figsize=(15, 10))
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
# Univariate Analysis contd..
# Categorical features
categorical_cols = ['Person_Home_Ownership', 'Loan_Intent', 'Loan_Grade', 'Cb_Person_Default_On_File']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate
# Numeric vs target
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Loan_Status', y=col, data=df)
    plt.title(f"{col} vs Loan_Status")
    plt.show()
```

```python
# Bivariate contd..
# Categorical vs target
# List of categorical columns to analyze
categorical_cols = ['Person_Home_Ownership', 
                    'Loan_Intent', 
                    'Loan_Grade', 
                    'Cb_Person_Default_On_File']

for col in categorical_cols:
    ct = pd.crosstab(df[col], df['Loan_Status'])
    ct.plot(kind="bar", figsize=(6,4))
    plt.title(f"Loan Status by {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.legend(title="Loan_Status")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation Analysis
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols + ['Loan_Status']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# outlier Detection
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```
